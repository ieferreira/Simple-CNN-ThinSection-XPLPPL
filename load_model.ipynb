### Load required libraries

```python
import numpy as np 
import matplotlib.pyplot as plt 
from tensorflow import keras
```

### Load pre-trained h5 model

### IMPORTANT: 
the model is stored in a series of .7z compressed files in the **model** folder. Please extract the **xplppl_modelv01.7z.001** file :c sorry... github file size limit

```python
try:
    model = keras.models.load_model('model/xplppl_modelv01.h5') # trained by Iv√°n Ferreira on a local dataset with a simple neural network
    print("Model loaded successfully")
except:
    print("\nModel NOT loaded")
    print("Did you extract the h5 file in the model folder?")
```

### Definition of model classes 

```python
# in the model 0 is xpl and 1 is ppl, so the next list suffices this classification

clases = ["XPL", "PPL"]
```

### Function to test images

```python
from PIL import Image



def probarUna(name):
    """
    takes the name of the filename of an image
    
    returns:
    image as np.array
    clasification as string (0 or 1 in this case XPL or PPL respectively)
    """
    imagen = Image.open(name)

    imagen = imagen.resize((400,400), Image.ANTIALIAS)

    imagen = np.array(imagen)

    import matplotlib.pyplot as plt


    test = np.expand_dims(imagen, axis=0)
    n = model.predict_classes(test)
    plt.imshow(imagen)
    return imagen, clases[int(n[0])]
```
